```python
def value_counts_to_dataframe(col):
    temp_df = df[col].value_counts().rename_axis(col).reset_index(name='counts')
    temp_df = temp_df.sort_values(by=[col])
    temp_df = temp_df.reset_index()
    temp_df.drop(columns=['index'],inplace=True)
    return temp_df
```

Modeling

```python
df = df.drop(['education'], axis=1)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns=['Gender', 'prevalentStroke','Heart_ stroke'], drop_first=True)
```

```python
df
```

Split to Train and Test

```python
from sklearn.model_selection import train_test_split

x = df.drop(['Heart_ stroke_yes'], axis=1)
y = df['Heart_ stroke_yes']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=18)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x_train)
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
x_test
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, fbeta_score, classification_report
```

```python
models = {
    'LR': LogisticRegression(),
    'KNN': KNeighborsClassifier(),
    'SVM': SVC(),
    'DT': DecisionTreeClassifier(),
    'RF': RandomForestClassifier(),
    'XGB': XGBClassifier()
}
```

```python
for name, model in models.items():
    print(f'Model Name: {name}')
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    print(f'Accuracy: {accuracy_score(y_test, y_pred)}')
    print(f'Recall: {recall_score(y_test, y_pred)}')
    print('-'*30)
```

```python
model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
grid={"C":np.logspace(-3,3,7), "penalty":["l2","l2"]}
logreg=LogisticRegression()
logreg=GridSearchCV(logreg,grid,cv=10)
logreg.fit(x_train,y_train)
```
